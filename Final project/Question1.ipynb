# Question 1 with K-Nearest Neighbors

### import neighbors to predict y

```python
import numpy as np
import pandas as pd
from sklearn import neighbors
```

```python
df = pd.read_csv('cancer_data.csv', encoding="ISO-8859-1", header=0)
df.head()
type(df)
```

```python
x=df.loc[:, 'radius': 'fractal_dimension' ]
```

```python
y=df.iloc[:,-1]
```

```python
k=15
```

```python
knn=neighbors.KNeighborsClassifier(n_neighbors=k, weights='distance')
```

```python
knn
```

```python
knn.fit(x,y)
```

```python
x
```

### predict y (cancer)

```python
pred=knn.predict([[14,14,88,566,1,0.08,0.06,0.04,0.18,0.05]])
# print(pred)
# print('%s %d' % (name, number))
print("Within these features we have %d person who has cancer with tumor " % pred)
```
